## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Step 3: Initial Inspection
print("Initial shape:", df.shape)
print("Columns:", df.columns.tolist())

# Step 4: Replace '?' with NaN
df.replace("?", np.nan, inplace=True)

# Step 5: Convert relevant columns to numeric
cols_to_convert = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Step 6: Handle Missing Values
# Fill numerical NaNs with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill categorical NaNs with mode
df.fillna(df.mode().iloc[0], inplace=True)

# Step 7: Basic Stats and Unique Value Checks
print("\nUnique values per column:")
print(df.nunique())
````

````python
import seaborn as sns

# Manually define horsepower ranges
def hp_category(hp):
    if hp <= 70:
        return 'Low (<=70)'
    elif hp <= 100:
        return 'Moderate (71-100)'
    elif hp <= 130:
        return 'High (101-130)'
    else:
        return 'Very High (>130)'

# Create new column
df['hp-range'] = df['horsepower'].apply(hp_category)

# Group by horsepower range and compute average price
avg_price_by_hp = df.groupby('hp-range')['price'].mean().reindex(['Low (<=70)', 'Moderate (71-100)', 'High (101-130)', 'Very High (>130)'])

# Use Seaborn Set2 color palette
colors = sns.color_palette("Set2", len(avg_price_by_hp))

# Plot bar chart
plt.figure(figsize=(8,5))
plt.bar(avg_price_by_hp.index, avg_price_by_hp.values, color=colors, edgecolor='black')
plt.title('Average Price by Horsepower Range')
plt.xlabel('Horsepower Range')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

````python
# Step 1: Confirm the required columns exist
if 'city-mpg' in df.columns and 'highway-mpg' in df.columns:

    # Step 2: Create avg-mpg column
    df['avg-mpg'] = (df['city-mpg'] + df['highway-mpg']) / 2

    # Step 3: Define mpg efficiency categories
    def mpg_range(mpg):
        if mpg <= 20:
            return 'Poor'
        elif mpg <= 25:
            return 'Average'
        elif mpg <= 30:
            return 'Good'
        else:
            return 'Excellent'

    # Step 4: Apply category labels
    df['mpg-bin'] = df['avg-mpg'].apply(mpg_range)

    # Step 5: Group by category and get average price
    avg_price_by_mpg = df.groupby('mpg-bin')['price'].mean().reindex(['Poor', 'Average', 'Good', 'Excellent'])

    # Step 6: Plot
    plt.figure(figsize=(10, 5))
    avg_price_by_mpg.plot(marker='o', linestyle='-', color='navy')
    plt.title('Average Price by MPG Efficiency Level')
    plt.xlabel('MPG Category')
    plt.ylabel('Average Price')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

else:
    print("Columns 'city-mpg' and/or 'highway-mpg' are missing in your DataFrame.")
````
